```python
import matplotlib.pyplot as plt
```

```python
P = [2, 4, 4, 4, 5, 5, 7, 9]
Q = [10, 13, 15, 14, 13, 16, 18, 21]
x = [1, 2, 3, 4, 5, 6, 7, 8]
```

```python
plt.scatter(x,P,color='r', marker='D')
plt.scatter(x,Q,color='b', marker='X')
plt.title('Correlation between P and Q')
plt.xlabe;('Range')
plt.ylabel('Value')
plt.legend((P,Q), loc='lower right')
plt.show()
```
