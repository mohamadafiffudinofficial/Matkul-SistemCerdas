```python
import numpy as np
```

```python
from sklearn import preprocessing
```

```python
sample_data = np.array([[2.1, -1.9, 5.5],
                       [-1.5, 2.4, 3.5],
                       [0.5, -7.9, 5.6],
                       [5.9, 2.3, -5.8]])
```

```python
sample_data
```

```python
sample_data.shape
```

```python
preprocessor = preprocessing.Binarizer(threshold=0.5)
```

```python
binarised_data = preprocessor.transform(sample_data)
```

```python
binarised_data
```

```python
preprocessor = preprocessing.MinMaxScaler(feature_range=(0,1))
```

```python
preprocessor.fit(sample_data)
```

```python
scaled_data = preprocessor.transform(sample_data)
```

```python
scaled_data
```

```python
scaled_data = preprocessor.fit_transform(sample_data)
```

```python
scaled_data
```

```python
l1_normalized_data = preprocessing.normalize(sample_data, norm='l1')
```

```python
l1_normalized_data
```

```python
l1_normalized_data = preprocessing.normalize(sample_data, norm='l2')
```

```python
l1_normalized_data
```

```python
label_kategori = ['senin', 'selasa', 'rabu', 'kamis', 'jumat', 'sabtu', 'minggu']
```

```python
encoder = preprocessing.LabelEncoder()
encoder.fit(label_kategori)
```

```python
print("\nLabel mapping:")
```

```python
for i, item in enumerate(encoder.classes_):
    print(item, '=', i)
```

```python
test_labels = ['senin', 'kamis', 'jumat']
```

```python
encoded_values = encoder.transform(test_labels)
print("Labels =", test_labels)
print("Encoded values =", list(encoded_values))
```
