```python
maknae = [
      'COVID-19 still becomes big threats for many countries, including Indonesia. Until now, Indonesia still struggle to reduce the positive cases while conducting the vaccination.'    
      'Even if some parts of society have got vaccinated, there are still many clusters of new cases.  Moreover, there is big threat that may happen since now people are preparing for Mudik Lebaran.'
      'In this case, it is better to make strict rules regarding mudik and other relevant activities. Government must make clear regulation and sanction for those who disobey the rules.'
      'In addition, potential area where people can gather must get special attention. There should be authorities to monitor the condition. When these are conducted well, event of Lebaran will not become big threat.'
]
maknae
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
vectorizer_X = vectorizer.fit_transform(maknae).todense()
vectorizer_X
```

```python
vectorizer.get_feature_names()
```

```python
from sklearn.metrics.pairwise import euclidean_distances

for i in range(len(vectorizer_X)):
    for j in range(i, len(vectorizer_X)):
        if i==j:
            continue
        jarak = euclidean_distances(vectorizer_X[i], vectorizer_X[j])
        print(f'Jarak dokumen {i+1} dan {j+1}: {jarak}')
```
